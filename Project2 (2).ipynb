```
import pandas as pd
import numpy as py
```

```
df = pd.read_csv('Indicators_of_Health_Insurance_Coverage_at_the_Time_of_Interview.csv')
# opening and reading the CSV
```

```
df
```

```
# splitting the df by phases
grouped_by_phase = df.groupby(df.Phase)
df_phase1 = grouped_by_phase.get_group(1)
df_phase2 = grouped_by_phase.get_group(2)
df_phase1
```

```
# looking only at VA
group_by_state_1 = df_phase1.groupby("State")
group_by_state_2 = df_phase2.groupby("State")
df_phase1_VA = group_by_state_1.get_group("Virginia")
df_phase2_VA = group_by_state_2.get_group("Virginia")
df_phase1_VA
```

```
import plotly
import plotly.express as px
```

```
px.bar(df_phase1_VA.sort_values(by = "Time Period"), y= "Value",
       color = "Time Period", hover_name= "Indicator",
       title = "VA percentages of Insurance Coverage (Phase 1)", labels = {"Value": "Percentage of Respondents", "Time Period Label": "Date"},hover_data = ["Time Period Label"])
```

```
px.bar(df_phase2_VA.sort_values(by = "Time Period Label"), y= "Value",
       color = "Time Period", hover_name= "Indicator",
       title = "VA percentages of Insurance Coverage (Phase 2)", labels = {"Value": "Percentage of Respondents", "Time Period Label": "Date"},hover_data = ["Time Period Label"])
```

```
grouped_by_state = df.groupby(df.State)
df_state = grouped_by_state.get_group("Virginia")
px.bar(df_state.sort_values(by = "Time Period Label"), y= "Value",
       color = "Time Period", hover_name= "Indicator",
       title = "VA percentages of Insurance Coverage (Phase 2)", labels = {"Value": "Percentage of Respondents", "Time Period Label": "Date"},hover_data = ["Time Period Label"])
```
